### 2. Feature engineering

**Extract all numerical features**

```python
# select numerical features
data_type = train.dtypes
num_list = list(train.columns.values[(data_type == 'int') | (data_type == 'float')])
train[num_list]
```

```python
plt.figure(figsize=(10,8))
sns.heatmap(train[num_list].corr(), vmin=-1, vmax=1)
plt.show()
```

**From the above figure, we can see cont11 and cont12 seem to have strong correlation. For accuracy we use value to decide. Let %97.5 be a threshold.**

```python
# select the upper triangle matrix of the correlation matrix
df_corr = train.corr()
df_corr_upper = pd.DataFrame(np.triu(df_corr, k=1)).abs()
```

```python
# column indices with abs correlation larger than the threshold.
dropcols_index = [j for j in range(len(df_corr_upper.columns)) if any(df_corr_upper.iloc[:,j] > 0.975)]
dropcols_index
```

**It turns out that the 13th column i.e., cont12 could be dropped based on the correlation threshold.**

```python
# create the list of redundent variables
dropcols = ['id', df_corr.columns[dropcols_index[0]]]
dropcols
```

```python
train = train.drop(dropcols, axis=1)
train.head()
```

**Encoding for the categorical data. Since categorical data columns are more than 100 and each of them have multiple categories, one-hot-encoding could lead to very large dimension of training data, which may affect training efficiency. So we use numerical values to replace the categorical letters (by using dictionary).**

```python
# Take a look at the categorical data
for col in train.columns:
    if col[:3] == 'cat':
        print(train[col].unique())
```

```python
Dict = {'A': 1, 'B': 2, 'C': 3, 'D': 4, 'E': 5, 'F': 6, 'G': 7, 'H': 8, 'I': 9, 'J': 10, 
        'K': 11, 'L': 12, 'M': 13, 'N': 14, 'O': 15, 'P': 16, 'Q': 17, 'R': 18, 'S': 19, 
        'T': 20, 'U': 21, 'V': 22, 'W': 23, 'X': 24, 'Y': 25, 'Z': 26}

# replace the letters by the corresponding values in a 26-base representation
# for example, 'AB' = 1*26 + 2 = 28.
for col in train.columns:
    if col[:3] == 'cat':
        train[col] = train[col].apply(lambda x: Dict[x] if len(x)==1 else Dict[x[0]]*10+Dict[x[1]])
train
```

**Check the distribution between train and test data**

### 3. Modeling and training

**We are using random forest for the regression here**

```python
# split data to train and validation
X_train, X_valid, y_train, y_valid = model_selection.train_test_split(train.iloc[:,:-1], train['loss'], test_size=0.2)
```

```python
print(X_train.shape, X_valid.shape, y_train.shape, y_valid.shape)
```

```python
# build random forest regressor
RFR = RandomForestRegressor(n_estimators=100, max_depth=5)
RFR.fit(X_train, y_train)
# RFR.predict(X_valid)-y_valid
# print (metrics.classification_report(y_test, y_pred))
# # Confusion matrix for training set (col1-actual incredible, col2-actual credible)
# metrics.confusion_matrix(RFR.predict(X_train), y_train)
```
